```python
categorical_cols=data.select_dtypes(include=['object']).columns
print('Categorical_col:',categorical_cols)
```

Observations:

The code lists all **categorical columns** (text/string type) in the dataset, such as `MSZoning`, `Neighborhood`, and `HouseStyle`. These columns need to be **encoded** before using them in machine learning models.


```python
numerical_cols=data.select_dtypes(include=['int64','float64']).columns
print('Numerical_cols:',numerical_cols)
```

Observations:

The code lists all **numerical columns** (integer and float types) in the dataset, such as `LotArea`, `GrLivArea`, and `SalePrice`. These columns can be used directly in models after handling **missing values** and **scaling** if needed.


```python
for col in numerical_cols:
    if data[col].nunique() < 10:
        print(f"Possibly Categorical (numeric): {col} -> {data[col].unique()}")
```

Observations:

Some numerical columns with fewer than 10 unique values (like `OverallQual`, `OverallCond`, `GarageCars`) are actually **categorical in nature**. These should be treated as **categorical features** during encoding and modeling.


```python
data=pd.get_dummies(data,drop_first=True)
```

Observations:

The code applies one-hot encoding to all categorical variables in the dataset and drops the first category to avoid multicollinearity

```python
data.isnull().sum().sort_values(ascending=False)
```

```python
data['MasVnrArea'].fillna(0,inplace=True)
```

Observations:

Missing values in `MasVnrArea` are filled with **0**, assuming no masonry veneer. This removes NaNs and makes the column ready for modeling.


```python
data.isnull().sum().sort_values(ascending=False)
```

Observations:

The code shows all columns **sorted by missing values**, highlighting which features still have NaNs. This helps decide the **appropriate imputation method** for each column.


```python
x=data.drop('SalePrice',axis=1)
y=data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

Observations:

Random Forest uses an ensemble of decision trees to capture complex relationships between features and SalePrice.The predicted values y_pred can now be evaluated using metrics like RMSE or R².

```python
from sklearn.metrics import mean_squared_error,r2_score
r2=r2_score(y_test,y_pred)
print(r2)
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse)
```

Observations:

The R²  score = 0.887  indicates that the Random Forest  model explains about 88.7% of the variance in house prices, which is very good.

The RMSE ≈ 29,469 means the model’s predictions are on average off by around $29,469 from the actual sale prices.
